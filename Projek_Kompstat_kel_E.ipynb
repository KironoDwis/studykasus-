```python
# Code untuk Soal No 01 diberikan sebagai Contoh dan Bonus
# Pelajari dengan baik

nilai_dosen = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
nilai_dosen = nilai_dosen.rename({'nilai': 'mean_nilai'}, axis=1) 
nilai_dosen = pd.merge(dosen, nilai_dosen, left_on='NIDN', right_index=True)
nilai_dosen = nilai_dosen[['NIDN', 'nama', 'mean_nilai']]
sb = nilai.groupby(nilai['NIDN']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1) 
nilai_dosen = nilai_dosen.set_index(['NIDN']) 
nilai_dosen = pd.concat([nilai_dosen, sb], axis=1)
nilai_dosen
```

3. Apakah terdapat perbedaan performa akademik antara mahasiswa yang berasal dari Madrasah/SMK dan SMU?

```python
# Code Soal No 03
nilai_mahasiswa = nilai.groupby(nilai['nim']).mean()[['nilai']]
nilai_mahasiswa = nilai_mahasiswa.rename({'nilai':'mean_nilai'}, axis=1)
nilai_mahasiswa = pd.merge(mahasiswa, nilai_mahasiswa, left_on='nim', right_index=True)
nilai_mahasiswa = nilai_mahasiswa[['nim', 'Asal_Sekolah', 'mean_nilai']]
sb_mah = nilai.groupby(nilai['nim']).std()[['nilai']]
sb_mah = sb_mah.rename({'nilai':'std_nilai'}, axis=1)
nilai_mahasiswa = nilai_mahasiswa.set_index(['nim'])
nilai_mahasiswa = pd.concat([nilai_mahasiswa, sb_mah], axis=1)
nilai_mahasiswa.pivot_table(index='Asal_Sekolah')
```

# Interpretasi 03:

Ya, terdapat perbedaan performa akademik antara mahasiswa yang berasal dari Madrasah/SMK dan SMU. Diperoleh rata rata nilai mahasiswa asal Madrasah yaitu 71.9375 , rata rata nilai mahasiswa asal SMK 85.9000, dan rata rata nilai mahasiswa SMU 66.4167. Dari perolehan tersebut, diketahui bahwa performa akademik mahasiswa yang berasal dari SMK dan Madrasah lebih baik daripada SMU jika dilihat dari nilai mean nya.

Adapun diperoleh juga nilai standar deviasi. Diketahui bahwa nilai standar deviasi mahasiswa asal SMK lebih kecil dari pada SMU dan Madrasah. Nilai standar deviasi yang rendah menggambarkan persebaran nilai yang kecil yang mengimplikasikan bahwa nilai nilai mahasiswa SMK semakin akurat terhadap mean. Sedangkan jika nilai standar deviasi tinggi, maka dapat diketahui terdapat nilai yang beragam (adanya kesenjangan). Sehingga dari ketiga asal sekolah di atas, diperoleh bahwa mahasiswa yang berasal dari SMK memiliki performa akademik terbaik dibanding Madrasah dan SMU.

```python
#Untuk matkul KAL01
nilai_kal01 = nilai.set_index(['id_mata_kuliah'])
nilai_kal01 = nilai_kal01.drop(['CS01','KAL02','STA02'])
nilai_kal01 = pd.merge(nilai_kal01, mahasiswa)
nilai_kal01 = nilai_kal01[['nim', 'nilai','Gender']]
mean_kal01 = nilai_kal01.groupby(nilai_kal01['Gender']).mean()[['nilai']]
mean_kal01 = mean_kal01.rename({'nilai': 'mean_KAL01'},axis=1)
sd_kal01 = nilai_kal01.groupby(nilai_kal01['Gender']).std()[['nilai']]
sd_kal01 = sd_kal01.rename({'nilai': 'std_KAL01'},axis=1)
nilai_kal01 = pd.concat([mean_kal01, sd_kal01], axis=1)
#Untuk matkul CS01
nilai_cs01 = nilai.set_index(['id_mata_kuliah'])
nilai_cs01 = nilai_cs01.drop(['KAL01','KAL02','STA02'])
nilai_cs01 = pd.merge(nilai_cs01, mahasiswa)
nilai_cs01 = nilai_cs01[['nim', 'nilai','Gender']]
mean_cs01 = nilai_cs01.groupby(nilai_cs01['Gender']).mean()[['nilai']]
mean_cs01 = mean_cs01.rename({'nilai': 'mean_CS01'},axis=1)
sd_cs01 = nilai_cs01.groupby(nilai_cs01['Gender']).std()[['nilai']]
sd_cs01 = sd_cs01.rename({'nilai': 'std_CS01'},axis=1)
nilai_cs01 = pd.concat([mean_cs01, sd_cs01], axis=1)
#Untuk matkul STA02
nilai_sta02 = nilai.set_index(['id_mata_kuliah'])
nilai_sta02 = nilai_sta02.drop(['CS01','KAL01','KAL02'])
nilai_sta02 = pd.merge(nilai_sta02, mahasiswa)
nilai_sta02 = nilai_sta02[['nim', 'nilai','Gender']]
mean_sta02 = nilai_sta02.groupby(nilai_sta02['Gender']).mean()[['nilai']]
mean_sta02 = mean_sta02.rename({'nilai': 'mean_STA02'},axis=1)
sd_sta02 = nilai_sta02.groupby(nilai_sta02['Gender']).std()[['nilai']]
sd_sta02 = sd_sta02.rename({'nilai': 'std_STA02'},axis=1)
nilai_sta02 = pd.concat([mean_sta02, sd_sta02], axis=1)
#Untuk matkul KAL02
nilai_kal02 = nilai.set_index(['id_mata_kuliah'])
nilai_kal02 = nilai_kal02.drop(['CS01','KAL01','STA02'])
nilai_kal02 = pd.merge(nilai_kal02, mahasiswa)
nilai_kal02 = nilai_kal02[['nim', 'nilai','Gender']]
mean_kal02 = nilai_kal02.groupby(nilai_kal02['Gender']).mean()[['nilai']]
mean_kal02 = mean_kal02.rename({'nilai': 'mean_KAL02'},axis=1)
sd_kal02 = nilai_kal02.groupby(nilai_kal02['Gender']).std()[['nilai']]
sd_kal02 = sd_kal02.rename({'nilai': 'std_KAL02'},axis=1)
nilai_kal02 = pd.concat([mean_kal02, sd_kal02], axis=1)
nilai_matkul = pd.concat([nilai_kal01, nilai_cs01, nilai_sta02, nilai_kal02], axis=1)
nilai_matkul
```

# Interpretasi 05

Apakah ada perbedaan prestasi akademik yang nampak di data antar angkatan mahasiswa?

dapat kita lihat bahwa terdapat perbedaan prestasi akademik antar angaktan mahasiswa dilihat dari mean(rata-tara)nilai antar angkatan dan standar deviasi nilai dari tiap angkatan


```python
#code soal no 5
prestasi_angkatan = nilai.groupby(nilai['nim']).mean()[['nilai']]
prestasi_angkatan = prestasi_angkatan.rename({'nilai': 'mean_nilai'}, axis=1)
prestasi_angkatan = pd.merge(mahasiswa, prestasi_angkatan, left_on='nim', right_index=True)
prestasi_angkatan = prestasi_angkatan[['nim', 'Angkatan', 'mean_nilai']]
sb = nilai.groupby(nilai['nim']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1) 
prestasi_angkatan = prestasi_angkatan.set_index(['nim']) 
prestasi_angkatan = pd.concat([prestasi_angkatan, sb], axis=1)
prestasi_angkatan.pivot_table(index='Angkatan')
```
